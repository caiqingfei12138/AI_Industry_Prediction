```python
import warnings
import pandas as pd

warnings.filterwarnings('ignore')

train = pd.read_excel('H:\\java\\python\\src\\machinelearning\\seconddata\\raw_data\\训练_20180117.xlsx')
test_A = pd.read_excel('H:\\java\\python\\src\\machinelearning\\seconddata\\raw_data\\测试A_20180117.xlsx')
test_B = pd.read_excel('H:\\java\\python\\src\\machinelearning\\seconddata\\raw_data\\测试B_20180117.xlsx')
# train_data = pd.concat([train,test_A,test_B]) # ignore_index=False
# data = train_data.drop('ID',axis=1)
target = train['Value']
train.drop('Value',axis=1,inplace=True)
missing_count = train.apply(lambda x: sum(x.isnull()))
```

```python
# missing_count.to_csv('H:\\java\\python\\src\\machinelearning\\seconddata\\preprocessing\\缺失值统计.csv')
```

```python
# 1，删除缺失较多的特征
more_missing = missing_count[missing_count>=799]
more_missing_index = list(more_missing.index)
train = train.drop(more_missing_index,axis=1)
```

```python
train_cols = train.columns
test_A = test_A[train_cols]
test_B = test_B[train_cols]
print(train.shape,test_A.shape,test_B.shape)
```

```python
train_data = pd.concat([train,test_A,test_B]) # ignore_index=False
data = train_data.drop('ID',axis=1)
```

```python
data.shape
```

```python
# 取出tool_ID
tool_cols = ['TOOL' , 'Tool', 'TOOL_ID','Tool (#1)','TOOL (#1)','TOOL (#2)',
                 'Tool (#2)','Tool (#3)','Tool (#4)','OPERATION_ID','Tool (#5)','TOOL (#3)']
gongxue_data = data[tool_cols]

data=data.drop(tool_cols,axis=1)
```

```python
data.shape
```

```python
# Imputer
from sklearn.preprocessing import Imputer
data = pd.DataFrame(Imputer().fit_transform(data))
```

```python
from sklearn.preprocessing import MinMaxScaler
#区间缩放，返回值为缩放到[0, 1]区间的数据
MinMaxScaler_data = pd.DataFrame(MinMaxScaler().fit_transform(data))
```

```python
# 方差分析

from sklearn.feature_selection import VarianceThreshold
 
#方差选择法，返回值为特征选择后的数据
#参数threshold为方差的阈值
VarianceThreshold_data = pd.DataFrame(VarianceThreshold(threshold=0.005).fit_transform(MinMaxScaler_data))  # 阈值需要调整
```

```python
# 相关系数计算
import scipy.stats as stats

corr = []
p_value = []
corr_data = VarianceThreshold_data.ix[0:799,:]
feature_cols = list(corr_data.columns)
y_col = target
for index_x in range(len(feature_cols)):
        r, p=stats.pearsonr(corr_data[index_x],target)
        corr.append(r)
        p_value.append(p)
        
import matplotlib.pyplot as plt

plt.plot(corr)
plt.show()
```

```python
s_corr = pd.Series(corr)
corr_filter = s_corr[abs(s_corr)>0.1]   # 阈值需要调整
```

```python
corr_cols=corr_filter.index
data = VarianceThreshold_data[corr_cols]
```

```python
data.head(5)
```

```python
# one-Hot Encode
one_hot = pd.get_dummies(gongxue_data, columns=tool_cols)
# onhot concat with VarianceThreshold_da
one_hot= one_hot.reset_index(drop=True)
```

```python
#数据拼接
# one_var_data=pd.concat([one_hot,data],axis=1,ignore_index=False)
one_var_data=pd.concat([one_hot,data],axis=1,ignore_index=False)
train = one_var_data.ix[0:799,:]
testA = one_var_data.ix[800:1099,:]
testB = one_var_data.ix[1100:,:]
```

```python
# XGBOOST
# build moudel
import xgboost as xgb
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error
from sklearn import ensemble, linear_model
from sklearn.linear_model.base import LinearRegression
from sklearn.neural_network.multilayer_perceptron import MLPRegressor
from sklearn.svm.classes import SVR
from sklearn.tree.tree import ExtraTreeRegressor, DecisionTreeRegressor
from sklearn.ensemble.bagging import BaggingRegressor
import warnings
warnings.filterwarnings('ignore')
mse = []
B = pd.DataFrame(np.zeros((300,10)))
sum_mse = 0.0

for index in range(10):
    X_train, X_test, y_train, y_test = train_test_split(train, target,test_size=0.1) 
    model=ensemble.RandomForestRegressor(n_estimators=25,
                                              criterion='mse',
                                              max_depth=14,
                                              min_samples_split=0.1,
                                              min_samples_leaf=2,
                                              min_weight_fraction_leaf=0.0,
                                              max_features=0.95,
                                              max_leaf_nodes=None,
                                              min_impurity_split=1e-07,
                                              bootstrap=True,
                                              oob_score=False,
                                              n_jobs=-1,
                                              random_state=None,
                                              verbose=0,
                                              warm_start=False)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse.append(mean_squared_error(y_test,y_pred))
    sum_mse += mean_squared_error(y_test,y_pred)
    predict = pd.DataFrame(model.predict(testA))
    B[index] = predict
print(pd.DataFrame(mse))
print('------------------------------------------------------------')
print(sum_mse/10)
```
