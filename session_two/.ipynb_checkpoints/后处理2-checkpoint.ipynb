```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# read target 

target= pd.read_excel('H:\\java\\python\\src\\machinelearning\\seconddata\\afterPreprocessing\\target.xlsx')
# target = pd.read_excel( 'H:\\java\\python\\src\\machinelearning\\seconddata\\raw_data\\Y_up.xlsx')
```

```python
plt.scatter(range(800),np.array(target))
plt.show( )
```

```python
def pos_value ( target ,ratio):
    size = target.size 
    sort_target = np.sort(np.array(target),axis=0)
    num = int(size*ratio)
    return sort_target[num] , sort_target[size-num]
```

```python
low, up = pos_value( target , 0.15)
low, up
```

```python
import warnings
warnings.filterwarnings('ignore')
# load features
features = pd.read_csv('H:\\java\\python\\src\\machinelearning\\seconddata\\afterPreprocessing\\data_id.csv')
# features = pd.read_csv('H:\\java\\python\\src\\machinelearning\\seconddata\\afterPreprocessing\\data_up.csv')
features.drop('ID',axis=1 , inplace=True)
train = features.ix[0:799,:]
testA = features.ix[800:1099,:]
```

```python
low_target = target [ target <= up ]
up_target  = target [ target >  up  ]
low_target=low_target.dropna()
up_target = up_target.dropna()
low_index = low_target.index
up_index  = up_target.index
target.ix[low_index,0] = 0
target.ix[up_index,0] = 1
```

```python
print(len(low_index) , len(up_index))
```

```python
plt.plot(target)
plt.show( )
```

```python
#training xgboost
import xgboost as xgb
import pandas as pd
dtrain = xgb.DMatrix(train,label=target)
dtest = xgb.DMatrix(testA)
```

```python
scale_pos_weight = float(len(target)-sum(target.values))/sum(target.values)
scale_pos_weight 
```

```python
params={'booster':'gbtree',
    'objective': 'binary:logistic',
    'scale_pos_weight':  5.72268908,
    'eval_metric': 'auc',
    'max_depth':6,
    'lambda':100,
    'subsample':0.65,
    'colsample_bytree':0.65,
    'eta': 0.002,
    'seed':1024,
    'nthread':12
}
```

```python
watchlist  = [(dtrain,'train')]
```

```python
#通过cv找最佳的nround
cv_log = xgb.cv(params,dtrain,num_boost_round=25000,nfold=5,metrics='auc',early_stopping_rounds=50,seed=1024)
bst_auc= cv_log['test-auc-mean'].max()
cv_log['nb'] = cv_log.index
cv_log.index = cv_log['test-auc-mean']
bst_nb = cv_log.nb.to_dict()[bst_auc]
```

```python
#train
watchlist  = [(dtrain,'train')]
model = xgb.train(params,dtrain,num_boost_round=bst_nb+50,evals=watchlist)
```

```python
#predict test set
test_y = model.predict(dtest)
predictions = [round(value) for value in test_y]
predictions = pd.DataFrame( predictions )
print (bst_nb,bst_auc)
print( predictions )
```
